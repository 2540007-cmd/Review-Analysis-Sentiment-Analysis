# Sentiment Analysis

```python
df1_clean = df1.copy()
```

```python
nltk.download("stopwords")
```

```python
stop_words = set(stopwords.words("english"))
```

```python
print(stop_words)
```

```python
negation_words = {
    "no", "not", "nor", "never",
    "don", "don't", "ain", "aren", "aren't",
    "couldn", "couldn't", "didn", "didn't",
    "doesn", "doesn't", "hadn", "hadn't",
    "hasn", "hasn't", "haven", "haven't",
    "isn", "isn't", "mightn", "mightn't",
    "mustn", "mustn't", "needn", "needn't",
    "shan", "shan't", "shouldn", "shouldn't",
    "wasn", "wasn't", "weren", "weren't",
    "won", "won't", "wouldn", "wouldn't"
}

stop_words = stop_words - negation_words
```

Negation words are some of the words that give the sentence its negitivity. For example: "<No> one should buy this.", "The product <dosn't> do what its supposed to do", etc.

```python
def clean_text(text):
    text = text.lower()

    #Remove links
    text = re.sub(r"http\S+|www\S+", "", text)

    #Remove usernames like @user
    text = re.sub(r"@\w+", "", text)

    #Removing hastags but keeping words
    text = re.sub(r"#", "", text)

    #Removing punctuations, numbers, and special characters
    text = re.sub(r"[^a-zA-Z\s']", "", text)

    #Removing extra spaces
    text = re.sub(r"\s+", " ", text).strip()

    #Split sentences into words
    words = text.split()

    #Remove stop words
    words = [word for word in words if word not in stop_words]

    #joining cleaned words back into sentences
    return " ".join(words)
```

This function takes one comment as input and processes it. The function does the following:
- Turns it all into Lowercase.
- Removes links. The re.sub() is used to find patterns and remove them.
  r"http\S+|www\S+" it finds links that start with http, or www. And \S+ means all non-space characters after that(http or www) must also be removed.
- Removes usernames.
  r"@\w+", here @ is removed and \w+ means the entire word that contains @. Example: "@Jhon", "@smith24", etc.
- Remove the hashtag symbol but keep the word.
- Removes punctuations, numbers and special characters by using this statment: r"[^a-zA-Z\s']"
  Here the ^ symbol means not. So the pattern means dont remove "a-z", "A-Z", "\s" meaning spaces and "'" meaing apostrophe.
- Removes extra spaces. r"\s+" not to be confused with \S+ replaces multiple spaces with one space.
  While .strip() removes spaces from the start and the end.
- Splits sentences into a lost of seaperate words where space differentiates between words.
- Removes stop_words. It means go through each word in words(list) and keep words that are not in stop_words.
- Lastly, it joins the contents of the list words with spaces turning it back into string and than returns it.
  return " ".join(words)

```python
df1_clean["clean_text"] = df1_clean["text"].apply(clean_text)
df_full_clean["clean_text"] = df_full_clean["text"].apply(clean_text)
```

.apply() function applies another function to every value in a pandas column.

# Data Visualization

```python
sentiment_counts = df_full_clean["sentiment"].value_counts()

plt.figure(figsize=(6, 4))
plt.bar(sentiment_counts.index, sentiment_counts.values)
plt.title("Overall Sentiment Distribution")
plt.xlabel("Sentiment")
plt.ylabel("Number of Comments")
plt.show()
```

Here we can see that on this post most comments were neutral.

```python
sentiment_counts = df_full_clean["sentiment"].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(
    sentiment_counts.values,
    labels=sentiment_counts.index,
    autopct="%1.1f%%"
)
plt.title("Sentiment Percentage")
plt.show()
```

```python
top_countries = df_full_clean["Country"].value_counts().head(10)

plt.figure(figsize=(10, 5))
plt.bar(top_countries.index, top_countries.values)
plt.title("Top 10 Countries by Number of Comments")
plt.xlabel("Country")
plt.ylabel("Number of Comments")
plt.xticks(rotation=45)
plt.show()
```

```python
top_country_names = df_full_clean["Country"].value_counts().head(10).index

top_country_data = df_full_clean[df_full_clean["Country"].isin(top_country_names)]

country_sentiment_top = pd.crosstab(
    top_country_data["Country"],
    top_country_data["sentiment"]
)

country_sentiment_top.plot(kind="bar", figsize=(12, 6))

plt.title("Sentiment Distribution in Top 10 Countries")
plt.xlabel("Country")
plt.ylabel("Number of Comments")
plt.xticks(rotation=45)
plt.show()
```

This graph shows how users from different countries reacted.

```python
age_sentiment = pd.crosstab(
    df_full_clean["Age of User"],
    df_full_clean["sentiment"]
)

#This compares age groups with sentiment.
#It counts how many positive, negative, and neutral comments exist in each age group.

age_sentiment.plot(kind="bar", figsize=(10, 5))

plt.title("Sentiment Distribution by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Number of Comments")
plt.xticks(rotation=45)
plt.show()
```

This graph shows how different age groups reacted.

```python
positive_words = " ".join(
    df_full_clean[df_full_clean["sentiment"] == "positive"]["clean_text"]
).split()

positive_counts = Counter(positive_words).most_common(20)

words = [item[0] for item in positive_counts]
counts = [item[1] for item in positive_counts]

plt.figure(figsize=(12, 5))
plt.bar(words, counts)
plt.title("Top 20 Words in Positive Comments")
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.show()
```

This graph shows the top 20 words used in positive comments.

```python
negative_words = " ".join(
    df_full_clean[df_full_clean["sentiment"] == "negative"]["clean_text"]
).split()

negative_counts = Counter(negative_words).most_common(20)

words = [item[0] for item in negative_counts]
counts = [item[1] for item in negative_counts]

plt.figure(figsize=(12, 5))
plt.bar(words, counts)
plt.title("Top 20 Words in Negative Comments")
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.show()
```

This graph shows the top 20 words used in negative comments.

```python
neutral_words = " ".join(
    df_full_clean[df_full_clean["sentiment"] == "neutral"]["clean_text"]
).split()

neutral_counts = Counter(neutral_words).most_common(20)

words = [item[0] for item in neutral_counts]
counts = [item[1] for item in neutral_counts]

plt.figure(figsize=(12, 5))
plt.bar(words, counts)
plt.title("Top 20 Words in Neutral Comments")
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.show()
```

This graph shows the top 20 words used in neutral comments.
